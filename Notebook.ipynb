## Part I : Introduction

### Object : 
The object that was selected for this project is the Travel Journal of a Continental Tour by Andrew MacGeorge (1810-1891), a journal created 1867-1868 which is kept in the University of Glasgow Archives and Special Collections among the Manuscript collections references under GB 247 MS Gen 13/1.

The record for this object is found on the University of Glasgow website for their collections : [GB 247 MS Gen 13/1](https://www.gla.ac.uk/collections/#/details?irn=266029&catType=C)

The contents of this journal include entries from various travel destinations across Europe and include accounts from the perspective of Andrew MacGeorge also detailing illustrations drawn in ink and photographs taken devoid of colour. Throughout my research processes I only took note of one illustration from McGeorge’s journal which contained colourized ink. 

```python
from PIL import Image
import matplotlib.pyplot as plt

files = [
    "Images/MS Gen 13 007 copy.jpg",
    "Images/MS Gen 13 087 copy.jpg",
]

images = [Image.open(f) for f in files]

plt.figure(figsize=(12, 4))  # width, height of the whole row

for i, img in enumerate(images):
    plt.subplot(1, len(images), i+1)
    plt.imshow(img)
    plt.axis("off")
    plt.title(files[i])

plt.show()
```

This contrast was the catalyst for this project: if colour is used to highlighting specific moments, how could colourization of MacGeorge’s sketches using Generative AI gain a new perspectives and insight into the cultural heritage of the object.
