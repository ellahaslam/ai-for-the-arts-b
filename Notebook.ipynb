### Object : 
The object that was selected for this project is the Travel Journal of a Continental Tour by Andrew MacGeorge (1810-1891), a journal created 1867-1868 which is kept in the University of Glasgow Archives and Special Collections among the Manuscript collections references under GB 247 MS Gen 13/1.

The record for this object is found on the University of Glasgow website for their collections : [GB 247 MS Gen 13/1](https://www.gla.ac.uk/collections/#/details?irn=266029&catType=C)

The contents of this journal include entries from various travel destinations across Europe and include accounts from the perspective of Andrew MacGeorge also detailing illustrations drawn in ink and photographs taken devoid of colour. Throughout my research processes I only took note of one illustration from McGeorge’s journal which contained colourized ink.

### AI Tools:

#### *DALL-E 3*

This project utilized DALL-E 3 (incorporated within OpenAI’s ChatGPT) a generative AI model which utilizes a diffusion model. This is a type of neural network that utilizes random noise as a starting point for creating images. The starting point in this project is MacGeorge’s sketches and uses the prompt (the sketches) as a reference to create an output that completes the users request. DALL-E 3 uses a separate text encoder to process the prompt digestible to the diffusion model. This model trained using millions of images correlating to text, this is to train the model on visual concepts so that the output corresponds to the correct input. This was the appropriate tool for this particular project  because of the ease of use and accuracy of interpretation of complex instructions, all to ensure the integrity of the object. Furthermore, when I was initially testing different generative AI, the output which produced the more accurate style was OpenAI’s DALL-E 3 in my opinion.

#### *Phase 2 : Contextual Assessment*

To correct the assumption made by the generative AI tool described in Phase 1, I incorporated McGeorge’s written entry as part of a new prompt. 

>     "At 10.45 we arrived at the Charing Cross Hotel (the train being late). It was a horrid dull wet day" (GB 247 MS Gen 13/1). 

This excerpt from the journal depicts a much different environment than what was rendered initially.


**New Prompt:**  Change the colouring to reflect this excerpt from the text: "At 10.45 we arrived at the Charing Cross Hotel (the train being late). It was a horrid dull wet day” (GB 247 MS Gen 13/1).

**Output:**

```python
from PIL import Image
import matplotlib.pyplot as plt

files = [
    "ImagesAIB/DALLE3-2-MS-Gen-13:007.png",
]

images = [Image.open(f) for f in files]

plt.figure(figsize=(12, 4))

for i, img in enumerate(images):
    plt.subplot(1, len(images), i+1)
    plt.imshow(img)
    plt.axis("off")
    plt.title(files[i])

plt.show()

# Assistance with this code was provided by Microsoft CoPilot
```

There was a dramatic shift in the re-colourization of the sketches. By providing context from the entry, the AI changed its saturated palette towards a desaturated one which rendered a more accurate depictions or a more authentic speculative restoration of the object which brings the audiences closer to the 19th century experience described by MacGeorge.  
However, during my brainstorming process I consulted my peers about how I would connect audiences to the object – we came to the agreement that re-colourization is not enough of a tool to bridge the historical object to modern audience. Therefore, the third phase was born to provide emotional connectivity and enhance the user experience by providing a tool where the story of the journal can come to life in a contemporary visual (mood board). 

#### *Phase 3 : Sentimental Interpretation*

This last phase explores the emotional aspect of the journals and how generative AI can produce sentiment from original text and images into a digital mood board. This will translate 19th century words and sketches into a contemporary visual which will provide a modern twist allowing audiences to engage with the object in a new way before reading actual entries. 
I acted as the director of the visuals providing historical context, while DALL-E 3 acted as an interpreter and tool to facilitate my whims. I bridged the two ideas together using Plutchik’s Wheel of Emotions, which maps human feelings into colour palettes. Also, acting as a basis for ensuring the rendered content was accurate.  




[**Plutchik’s Wheel of Emotions**](https://www.6seconds.org/2025/02/06/plutchik-wheel-emotions/)

```python
from PIL import Image
import matplotlib.pyplot as plt

files = [
    "ImagesAIB/Plutchik-Model-6seconds.org.jpg",
]

images = [Image.open(f) for f in files]

plt.figure(figsize=(12, 4))

for i, img in enumerate(images):
    plt.subplot(1, len(images), i+1)
    plt.imshow(img)
    plt.axis("off")
    plt.title(files[i])

plt.show()

# Assistance with this code was provided by Microsoft CoPilot
```

**New Prompt:** Generate a mood board (as a contemporary visual) based on the sentiments provided from MacGeorge’s entry “At 10.45 we arrived at the Charing Cross Hotel (the train being late). It was a horrid dull wet day" (GB 247 MS Gen 13/1).

**Output:**

```python
from PIL import Image
import matplotlib.pyplot as plt

files = [
    "ImagesAIB/DALLE3-Generated-Mood-Board.png",
]

images = [Image.open(f) for f in files]

plt.figure(figsize=(12, 4))

for i, img in enumerate(images):
    plt.subplot(1, len(images), i+1)
    plt.imshow(img)
    plt.axis("off")
    plt.title(files[i])

plt.show()

# Assistance with this code was provided by Microsoft CoPilot
```

In this output you can use Plutchik’s Wheel as a basis for understanding colour as explained in the AI Generated Mood Board. For example, the dull blues and greens facilitating McGeorge’s feelings of apprehension and disapproval. 

This demonstrates an output which provides a nuanced experience of the object through design languages (such as colour, typography, saturation, etc) which encourages a new perspective of the object. This idea may bridge together the historical gap by providing a tool where audiences can relate to a specific feeling instead of trying to understand a black-and-white sketch from the 19th century. The use of generative AI provides a nuanced take on cultural heritage– taking the historic object and turning it into a dynamic and sentimental experience relevant to modern culture.
